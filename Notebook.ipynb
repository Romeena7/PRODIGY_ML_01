```python
data = pd.read_csv("House_Price_India.csv")
```

```python
#for na values
data.dropna(inplace = True)
#for duplicated values
data.drop_duplicates(inplace = True)
```

```python
import matplotlib.pyplot as plt
```

```python
data.groupby("condition of the house")["Price"].mean().sort_values(ascending=False).plot(kind="bar")
plt.title("Condition of the house VS Price")
plt.ylabel("Mean Price")
plt.xlabel("Condition of the house")
plt.show()
```

```python
data.columns
```

```python
data[["number of bedrooms","number of bathrooms"]]
```

```python
X = data[['number of bedrooms', 'number of bathrooms','living area','condition of the house']]
y = data[["Price"]]
```

```python
X
```

```python
y
```

```python
data.shape[0] * 0.2
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,)
```

```python
y_train = y_train.values.ravel()
```

```python
X_train.shape
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.tree import DecisionTreeRegressor
param_grid = {
    "criterion" : ["squared_error", "friedman_mse", "absolute_error"],
    "splitter" : ["best", "random"],
    "max_depth" : [None, 10, 15],
    "min_samples_split" : [2,5],
    "min_samples_leaf" : [1,2]
}
```

```python
tree_model = DecisionTreeRegressor()
```

```python
grid_tree = GridSearchCV(estimator=tree_model, param_grid = param_grid, cv=3)
```

```python
grid_tree.fit(X_train, y_train)
```

```python
grid_tree.best_params_
```

```python
tree_preds = grid_tree.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, tree_preds)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
predslr = lr.predict(X_test)
```

```python
mean_absolute_error(y_test, predslr)
```

```python
from sklearn.ensemble import RandomForestRegressor  
rfrmodel =  RandomForestRegressor()
```

```python
param_gridrfr = {
    "max_depth" : [5,10,15],
    "n_estimators" : [2,3,4,5,6,7,8,9,10]
}
```

```python
gridrfr = GridSearchCV(rfrmodel, param_gridrfr)
```

```python
gridrfr =  GridSearchCV(rfrmodel, param_gridrfr)
```

```python
gridrfr.fit(X_train, y_train.ravel())
```

```python
gridrfr.best_params_
```

```python
rfrpredictions = gridrfr.predict(X_test)
```

```python
mean_absolute_error(y_test, rfrpredictions)
```
